```python
import sys

import time

from pyclick.click_models.Evaluation import LogLikelihood, Perplexity
from pyclick.click_models.UBM import UBM
from pyclick.click_models.DBN import DBN
from pyclick.click_models.SDBN import SDBN
from pyclick.click_models.DCM import DCM
from pyclick.click_models.CCM import CCM
from pyclick.click_models.CTR import DCTR, RCTR, GCTR
from pyclick.click_models.CM import CM
from pyclick.click_models.PBM import PBM
# from pyclick.click_models.alternative_inference.BBM import BBM
# from pyclick.click_models.alternative_inference.UBMProbit import UBMProbit
from pyclick.utils.Utils import Utils
from pyclick.utils.YandexRelPredChallengeParser import YandexRelPredChallengeParser
from pyclick.utils.YandexPersonalizedChallengeParser import YandexPersonalizedChallengeParser
```

```python
click_model = globals()['CM']()
search_sessions_path = 'examples/data/YandexPersonalizedChallenge'
search_sessions_num = 10000
```

```python
search_sessions = YandexPersonalizedChallengeParser().parse(search_sessions_path, search_sessions_num)

train_test_split = int(len(search_sessions) * 0.95)
train_sessions = search_sessions[:train_test_split]
train_queries = Utils.get_unique_queries(train_sessions)

test_sessions = Utils.filter_sessions(search_sessions[train_test_split:], train_queries)
test_queries = Utils.get_unique_queries(test_sessions)
print "-------------------------------"
print "Training on %d search sessions (%d unique queries)." % (len(train_sessions), len(train_queries))
print "-------------------------------"
start = time.time()
click_model.train(train_sessions)
end = time.time()
print "\tTrained %s click model in %i secs:\n%r" % (click_model.__class__.__name__, end - start, click_model)
```

```python
import operator
sessions_filename = 'examples/data/YandexPersonalizedChallenge'
sessions_file = open(sessions_filename, "r")

query_dict = {}
for line in sessions_file:
    if len(query_dict) >= 100000:
        break
    entry_array = line.strip().split("\t")

    if len(entry_array) == 4 and entry_array[1] == 'M':
        pass

    elif len(entry_array) >= 7 and (entry_array[2] == 'Q' or entry_array[2] == 'T'):
        task = entry_array[0]
        serp = entry_array[3]
        query = entry_array[4]
        if query not in query_dict.keys():
            query_dict[query] = 1
            continue
        query_dict[query] += 1
    elif len(entry_array) == 5 and entry_array[2] == 'C':
        if entry_array[0] == task and entry_array[3] == serp:
            query_dict[query] += 1

    else:
        print 'Unknown data format: %s' % line
        continue

sessions_file.close()
x = query_dict
sorted_x = sorted(x.items(), key=operator.itemgetter(1))

f = open('new_train', 'wb')
new_query = [item[0] for item in sorted_x[-10:]]

qq = 0 
sessions_file = open(sessions_filename, "r")
for line in sessions_file:
    qq += 1
    if qq >= 100000:
        break
    entry_array = line.strip().split("\t")

    if len(entry_array) == 4 and entry_array[1] == 'M':
        pass

    elif len(entry_array) >= 7 and (entry_array[2] == 'Q' or entry_array[2] == 'T'):
        task = entry_array[0]
        serp = entry_array[3]
        query = entry_array[4]
        if query not in new_query:
            task = -1
            serp = -1
            continue
        f.write(line)
        print entry_array
    elif len(entry_array) == 5 and entry_array[2] == 'C':
        if entry_array[0] == task and entry_array[3] == serp:
            f.write(line)

    else:
        print 'Unknown data format: %s' % line
        continue
f.close()
```
